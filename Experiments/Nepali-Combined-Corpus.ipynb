**Oscar Data and News Dataset are not part of the research topic. But we use them here for better tokenizer and may also use them for pretraining as well.**

```python
%%time
## Reading Nepali News Data Sets
from tqdm.auto import tqdm

fp=[]
for dirname, _, filenames in os.walk("/kaggle/input/nepali-news-dataset-large/nepali_news_dataset_20_categories_large/nepali_news_dataset_20_categories_large"):
    fp += [os.path.join(dirname, filename) for filename in filenames if 'txt' in filename]

for dirname, _, filenames in os.walk("/kaggle/input/nepalinewsdataset"):
    fp += [os.path.join(dirname, filename) for filename in filenames if 'txt' in filename]
    
    
combined_news_data = []
for fn in tqdm(fp):
    with open(fn, 'rb') as f:
        lines = f.read()
        if len(lines) < 5:
            continue
        try:
            encoding = chardet.detect(lines)['encoding']
            lines = lines.decode(encoding)
            
            lines = lines.split("\r\n")
            combined_news_data+=lines
        except Exception as e:
            print("Couldn't Read file",fn,encoding, e)

            
combined_news_data = pd.DataFrame({'text': combined_news_data})
combined_news_data.drop(combined_news_data[combined_news_data['text'] == ''].index,inplace=True)

combined_news_data = datasets.Dataset.from_pandas(combined_news_data)
print("News Dataset")
print(combined_news_data.shape)
print(combined_news_data[:5])
gc.collect()
```

```python
%%time
fp = []
for dirname, _, filenames in os.walk("/kaggle/input/nepali-wikipedia-articles"):
    fp += [os.path.join(dirname, filename) for filename in filenames if 'txt' in filename]
    
for dirname, _, filenames in os.walk("/kaggle/input/urdunepali-parallel-corpus/"):
    fp += [os.path.join(dirname, filename) for filename in filenames if 'txt' in filename]
    
wiki_and_unp_corpus = []
for fn in tqdm(fp):
    with open(fn, 'rb') as f:
        lines = f.read()
        if len(lines) < 5:
            continue
        try:
            encoding = chardet.detect(lines)['encoding']
            if not encoding:  
                encoding = 'utf-8'
                
            lines = lines.decode(encoding)
            lines = lines.split("\n")
            
            wiki_and_unp_corpus+=lines
        except Exception as e:
            print("Couldn't Read file",fn,encoding, e)
            
wiki_and_unp_corpus = pd.DataFrame({'text': wiki_and_unp_corpus})
wiki_and_unp_corpus.drop(wiki_and_unp_corpus[wiki_and_unp_corpus['text'] == ''].index,inplace=True)

wiki_and_unp_corpus = datasets.Dataset.from_pandas(wiki_and_unp_corpus)
print("Wikipedia and UrduNepali-Parallel Dataset")
print(wiki_and_unp_corpus.shape)
print(wiki_and_unp_corpus[:1])
del fp
gc.collect()
```

```python
%%time
# ## Citation for this data set: https://inria.hal.science/hal-02148693

oscar_data = datasets.Dataset.load_from_disk('/kaggle/input/cleaning-oscar-nepali-dataset/NepaliCorpus')
print("Oscar Nepali Dataset")
print(oscar_data[:5])
gc.collect()
oscar_data
```

```python
from datasets import concatenate_datasets

print("Combining All the datasets")
data = concatenate_datasets([tweet_data,combined_news_data,wiki_and_unp_corpus ,oscar_data,datasets.Dataset.from_pandas(cov_news_data)])
data = data.remove_columns(['__index_level_0__'])

print("saving to disk")
data.save_to_disk("NepaliCombinedCorpus")
print("loading from disk") # this is just a nuance specific for kaggle :D :D :D ;(
data = datasets.Dataset.load_from_disk("NepaliCombinedCorpus")
gc.collect()
data
```

```python
%%time
print("Cleaning data: removing non nepali character")
data = data.map(removeCharacters,
                num_proc=4,
                batched=True
               )
data = data.filter(lambda x: len(x['text']) >= 50)
data
```
